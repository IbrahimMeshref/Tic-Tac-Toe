```python
import numpy as np
matrix=np.array([[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']])

lst_player=['O','X']
```

```python
def check():
    lis_win=[len(set(list(matrix[0]))),
    len(set(list(matrix[1]))),
    len(set(list(matrix[2]))),
    len(set(list(matrix[0:,0]))),
    len(set(list(matrix[0:,1]))),
    len(set(list(matrix[0:,2]))),
    len(set(list(list(matrix[0,0])+list(matrix[1,1])+list(matrix[2,2])))),
    len(set(list(list(matrix[2,0])+list(matrix[1,1])+list(matrix[0,2])))),
            ]
    return lis_win
```

```python
class player:
    turn=True
    end=True
    count=0
    def __init__(self,end,count):
        self.end=end
        self.count=count
        matrix.fill(' ')
    def display():
        for i in matrix:
            print('  ___'+' ___ '+'___')
            print(" | "+" | ".join(i)+" | ")



    def swap_play(self):

        print('It\'s the player\'s turn '+lst_player[self.turn])
        try:
            row=int(input('Enter Row place choose(1 or 2 or 3): '))
            column=int(input('Enter Column place choose(1 or 2 or 3): '))
            if((row>=1 and row<=3) and (column>=1 and column<=3) ):
                if(matrix[row-1,column-1]==' '):
                    matrix[row-1,column-1]=lst_player[self.turn]
                    result=check()
                    self.count+=1
                    ll=[i for i in result]

                    if ((row==2 or column==2) and row!=column):
                        if(ll[row-1]==1 or ll[column-1+3]==1 ):
                            print(lst_player[self.turn]+' is winner ðŸ¥³')
                            self.end=False

                    elif(column==row and row==2):
                        if(ll[-2]==1 or ll[row-1]==1 or ll[column-1+3]==1 or ll[-1]==1 ):
                            print(lst_player[self.turn]+' is winner ðŸ¥³')
                            self.end=False

                    else:
                        if(column==row):
                            if(ll[-2]==1 or ll[row-1]==1 or ll[column-1+3]==1 ):
                                print(lst_player[self.turn]+' is winner ðŸ¥³')
                                self.end=False
                        else:
                            if(ll[-1]==1 or ll[row-1]==1 or ll[column-1+3]==1 ):
                                print(lst_player[self.turn]+' is winner ðŸ¥³')
                                self.end=False


                    if(self.count==9):
                        print('DRAW ðŸ™‚')
                        self.end=False

                    self.turn=not self.turn
                    player.display()


                else:
                    print("!!!!!!!!This place has already been selected. Please choose another place,")
            else:
                print('!!!!! please enter row and colum between 1 and 3')
        except:
            print('Invalid input')
```

```python
print('****if want any choose place use Row and Colmun****')
print('ps:""You will enter a row the first time and the second time you will enter a column when play""')
game=player(True,0)
player.display()
play='y'
while(play=='y'):
    while(game.end):
        game.swap_play()
        print()
    play=input('If want play again (enter y) no (enter n) ')
    if(play=='y'):
        game=player(True,0)
321
```
